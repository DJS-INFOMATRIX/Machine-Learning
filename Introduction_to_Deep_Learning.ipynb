Deep Learning is a subset of Machine Learning. It mainly focuses on designing neural networks to solve the different machine learning problems. The idea of neural networks comes from our brains. Our brain is made up of cells known as neurons. These cells not only help transmit the signals throughout our nervous system. But, is also able to process information coming from our sensory organs and then work on that information correctly.

This, motivated scientists to come up with a structure similar to the neurons such that it is able learn from the information. These are known as a perceptron.

image.png

**Components of an Artificial neuron :**

**1) Inputs :** It is the different features that affect the decision taken by the neuron.

**2) Weights :** It is the priority of sorts that the feature has when determining the output.

**3) Weighted sum :** It is basically the sum of the products of the weights and the input values.

**4) Bias :** This is like a partial opinion that is introduced in the model.

**5) Activation function :** These are generally non-linear functions. Non-linear functions are used in neural networks because patterns occuring in real world data aren't always linear related thus we make use of different non-linear functions. This functions main aim is to identify the underlying pattern in the data and then produce an output based on what the activation fucntion determines.

A neural network is the combination of many perceptron in mulitple layers. The perceptron in one layer is connected to all the perceptrons in the next layer. This is why the neural network is called a Feed-Forward network, since the output of the first layer is the input of the next layer.

The loss function in a neural network is a mathematical function that evaluates the error / loss between the predicted value and the actual value.  The lower the value of loss the better the model. The model is improved with the help of backpropagation.

Backpropagation is the method by which the loss is calculated and is propagated backwards into the layers, adjusting the values of the weights so that the model is adjusted to predict more accurately.

Gradient descent is the process by which the model improves its perfomance such that its loss is reduced. The model has a learning rate which determines how fast it learns from the dataset.

Setting a high learning rate isn't always the best since there is a possibility that the model overshoots and misses the global minima.

Conversely, setting the learning rate to low will make the model learn very slowly and there is a possibility that the model is stuck at a local minima.

```python
# We can have special type of tensors such as all ones or zeros or identity matrices

x=tf.ones((3,3),dtype='int32')
print(x)

x=tf.zeros((3,3),dtype='int32')
print(x)

x=tf.eye(3,dtype='float16')   # eye() is used for identity matrix
print(x)
```

```python
# We can generate a tensor of random values based on the mean and the standard deviation.
x=tf.random.normal((3,3), mean=5, stddev=3)
print(x)

# We can generate a tensor of random values based on the min and max values these random numbers must lie in.
x=tf.random.uniform((3,3),minval=0,maxval=5)
print(x)
```

```python
# We can make a tensor in the same way of the range function used while traversing lists.
x=tf.range(9)
print(x)

# We can even specify starting and ending indices and the steps.
x=tf.range(start=1,limit=10,delta=2)
print(x)
```

```python
# We can change the datatype of the elements of an existing tensor using the cast()
x=tf.cast(x,dtype='float32')
print(x)
```

# Mathematical Operations

```python
x=tf.constant([1,2,3])
y=tf.constant([4,5,6])
```

```python
#Addition can be done using the add() function or the '+' operator
z=tf.add(x,y)
print(f"Addition : {z}")

z=x+y
print(f"Addition : {z}")
```

```python
#Subtraction can be done using the subtract() function or the '-' operator
z=tf.subtract(x,y)
print(f"Subtraction : {z}")

z=x-y
print(f"Subtraction : {z}")
```

```python
# Element-wise division can be done using the divide() function or the '/' operator
z=tf.divide(x,y)
print(f"Element-wise Division : {z}")

z=x/y
print(f"Element-wise Division : {z}")
```

```python
# Element-wise multiplication can be done using the multiply() function or the '*' operator
z=tf.multiply(x,y)
print(f"Element-wise Multiplication : {z}")

z=x*y
print(f"Element-wise Multiplication : {z}")
```

```python
# We can raise the elements of the tensor to a power using '**' operator
z=x**5
print(f"Power of each element: {z}")
```

```python
# For dot product between two tensors we use the tensordot() function
z=tf.tensordot(x,y,axes=1)    # axes=1 is required so as to determine the dot product wrt the columns of the tensor
print(f"Dot Product: {z}")
```

```python
# For matrix multiplication of tensors we use the matmul() function or '@' operator
a=tf.random.normal((2,3))
b=tf.random.normal((3,4))

print(f"A\n{a}")
print(f"B\n{b}")
c=tf.matmul(a,b)
print(f"Matrix Multiplication:\n{c}")

c=a @ b
print(f"Matrix Multiplication:\n{c}")
```

# Basics of Neural Networks on MNIST dataset

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.datasets import mnist
```

```python
(x_train,y_train),(x_test,y_test)=mnist.load_data()
```

```python
print(x_train.shape)
'''
The shape function is used to give the dimensions of the tensor, here we have a 3D tensor. In terms of ML and image processing, it, means the following:
'60000' means that there are 60000 data points in the mnist dataset.
'28,28' means that the images are of the dimension 28px X 28px.
'''

print(y_train.shape)
```

```python
'''
Now we reshape the images such that the height and width of the image are converted into a 1-D array.
This will make the utlization of the data in the connected layers of the neural network better.
'''

x_train=x_train.reshape(-1,28*28).astype('float32')/255.0
# The -1 signifies the number of datapoints should remain the same (i.e. 60,000) and the dimension of the 1-D array will be 784
# The default datatype of reshape is float64 that is converted to float32 with the help of the astype function.
# Also, dividing by 255 is done so as to normalize the values of the array (i.e. the values will range from 0 to 1) we take 255 since the hex value of any pixel lies between 0-255.

x_test=x_test.reshape(-1,28*28).astype('float32')/255.0
```

```python
x_train[0]
```

```python
'''
After preparing the data, now we prepare the neural network layers. That is done with the help of keras.
The problem with the sequential API is that it only works on a single input. Therefore, it is convenient to use but it isn't flexible.
We define the neural net model, by specifying the number of nodes in each layer of the neural net as well as defining the activation fucntion that will be used in the layer.
'''

model=keras.Sequential(
    [
      layers.Dense(512,activation='relu'),
      layers.Dense(256,activation='relu'),
      layers.Dense(10,activation='softmax')
    ]
)
```

There are different scenarios when an activation fumction is used in the output layer and which activation function is used when. The following are the cases:

1) Regression problems : When the neural network is used for regression we do not use any activation function, since we want to obtain the calue that was predicted by the neural network.

2) Binomial classification : When the neural network is used for binomial classification, we make use of the the **sigmoid function** in the output layer so that we get the probability of either of the case occuring, since the sigmoid function sqaushes the output values between 0 and 1.

3) Multi-class classification : When the neural network is used for multi-class classification, we make use of the **softmax function** which converts the output into the probability distribution over the classes.

Activation functions are crucial components of neural networks. They introduce non-linearities to the model, enabling it to learn complex mappings from inputs to outputs. Here are some common activation functions and their typical use cases:

**Sigmoid Activation Function:**

*Formula:*
$$
\sigma(x) = \frac{1}{1 + e^{-x}}
$$

*Range:* (0, 1)

*Use Case:* Often used in the output layer of binary classification models where the goal is to produce probabilities. It squashes the input values between 0 and 1.

****

**Hyperbolic Tangent (tanh) Activation Function:**

*Formula:*
$$
(\text{tanh}(x) = \frac{e^{2x} - 1}{e^{2x} + 1})
$$

*Range:* (-1, 1)

*Use Case:* Similar to the sigmoid, but with a range of (-1, 1). Commonly used in hidden layers of neural networks.

****

**Rectified Linear Unit (ReLU):**

*Formula:*
$$
f(x)=max(0,x)
$$

*Range:* [0,∞)

*Use Case:* One of the most popular activation functions. It introduces non-linearity and is computationally efficient. Commonly used in hidden layers of deep neural networks.

****

**Leaky Rectified Linear Unit (Leaky ReLU):**

*Formula:*
$$
f(x)=max(αx,x)
$$
where α is a small positive constant (e.g., 0.01)

*Range:* (−∞,∞)

*Use Case:* Addresses the "dying ReLU" problem by allowing a small, non-zero gradient for negative inputs. Used in situations where ReLU might result in dead neurons.

****

**Parametric Rectified Linear Unit (PReLU):**

*Formula:*
$$
f(x)=max(αx,x)
$$
α is a learnable parameter.

*Range:* (−∞,∞)

*Use Case:* Similar to Leaky ReLU but with α being a trainable parameter. It allows the network to learn the optimal slope for negative values.

****

**Exponential Linear Unit (ELU):**

*Formula:*

For x >= 0
$$
f(x)=x
$$

For x < 0
$$
f(x)=α(ex−1)
$$
where α is a hyperparameter.


*Range:* (-α, ∞)

*Use Case:* Addresses the vanishing gradient problem and has a smoother transition for negative inputs compared to ReLU.

****

**Softmax Activation Function:**

*Formula:*
$$
f(x)_i = \frac{e^{x_i}}{\sum_{j} e^{x_j}}
$$
for each element i in the vector x.

*Range:* (0, 1) for each element, and the elements sum to 1.

*Use Case:* Commonly used in the output layer of multi-class classification models. It converts raw logits into probabilities, ensuring that the sum of probabilities across all classes is equal to 1.

```python
#We compile this model, compiling is basically setting guidlines for fitting the data in the neural net

model.compile(
    loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    metrics=['accuracy']
)
```

**Understanding the different parameters in the compile method**

**1) loss**: It is the parameter used to set a loss function for the model.
Loss functions are those functions which determines how well the model is working which is determined by comparing the predicted value to the actual value.

In this case we have chosen the "SparseCategoricalCrossentropy" function. Cross entropy is the measure of difference between two probability distribution.
But in case of ML and deep learning, cross entropy is used as a loss function to quantify how well the predicted probability distribution fares against the true distribution.


There are two types of cross entropies, binary and categorical.


Binary is used in case of classification between two values, while categorical is used when there are multiple classes.
Since we need to classify which number the shown number is (i.e. 10 possible option), we use categorical cross entropy.

There are two types categorical cross entropies:

1) Sparse Categotical Cross entropy which is used when the target labels are provided as integers (in this case from 0 to 9).
    
2) Categotical Cross entropy which is used when the target labels are represented in the form of binary vectors.

3) Binary cross entropy is the loss function that is used for binary classification.

Therefore, SparseCategoricalCrossentropy is more useful in this case.
for_logits is a parameter which takes boolean values, it specifies whether the output must be considered as raw logits or probabilites.True sets that the output will be raw logits

****

**2) optimizers:** We know, that the neural network learns from the previous training data points and adjusts itself to increase its accuracy. This is done with the help of the optimizer. So the optimizer parameter updates the weights based on the gradients that are computed by the loss function. Here we have used Adam optimizer. It is a popular optimization algorithm which adapts the learning rate for each parameter. The parameter learning rate is used to set the initial learning rate of the algorithm. In this case the initial learning rate is 0.001.

****

**3)metrics:** It is used to evaluate the performance of the model while training and testing, in this case we evaluate the performance of the model based on the accuracy of the model.

```python
history=model.fit(x_train,y_train,batch_size=32,epochs=5,verbose=2)
model.evaluate(x_test,y_test,batch_size=32,verbose=2)
```

**Understanding the parameters from the fit and evaluate methods**

The fit function is used to train the model on the basis of the data. Whereas, the evaluate function is used to evaluate the accuracy of the neural network when a testing dataset is put through the neural net.

The x_train,y_train and x_test,y_test are the datasets that are passed into the function upon which the model is fit and evaluated respectively.

**batch_size :** It is used to set the number of datapoints that will be put into the model for training or testing at a time. The weights of the neural net are updated based on the results obtained by each batch, such that the weights are altered in a way that the accuracy of the neural network is increased.

**epochs :** The entire dataset is passed through the model multiple times so that weights are adjusted such that the accuracy of the model is maximized. An epoch is the count for the number of times the entire dataset was passed throught the model. This number can be set using the epochs parameter in the fit function.

**verbose :** It is used to set the amount of information is displayed during the training and testing of the model, where 2 is more amount of information and 0 is less amount of information.
